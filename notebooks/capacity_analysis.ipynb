```python
import pandas as pd
import numpy as np
import os

# Try the CSV from the current working directory first
csv_path = "HHS_Unaccompanied_Alien_Children_Program.csv"
if not os.path.exists(csv_path):
    csv_path = os.path.join(os.getcwd(), "HHS_Unaccompanied_Alien_Children_Program.csv")

if not os.path.exists(csv_path):
    raise FileNotFoundError("Could not find HHS_Unaccompanied_Alien_Children_Program.csv")

# Load the CSV file
try:
    df = pd.read_csv(csv_path)
except Exception as e:
    raise RuntimeError(f"Failed to load the CSV file: {e}")

# Clean and standardize column names
new_columns = {
    "Date": "date",
    "Children apprehended and placed in CBP custody*": "children_apprehended_and_placed_in_cbp_custody",
    "Children in CBP custody": "cbp_custody",
    "Children transferred out of CBP custody": "transferred_out_of_cbp_custody",
    "Children in HHS Care": "hhs_care",
    "Children discharged from HHS Care": "discharged_from_hhs_care"
}
df = df.rename(columns=new_columns)

# Drop rows with missing dates
df = df.dropna(subset=["date"])

# Convert the date column to datetime and sort chronologically
try:
    df["date"] = pd.to_datetime(df["date"], format="%B %d, %Y", errors="coerce")
except Exception:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Remove invalid dates and sort from oldest to newest
df = df.dropna(subset=["date"]).sort_values("date").reset_index(drop=True)

# Clean the HHS care column by removing commas and converting to numeric
# Using float here keeps it compatible with more pandas versions
if "hhs_care" in df.columns:
    df["hhs_care"] = pd.to_numeric(
        df["hhs_care"].astype(str).str.replace(",", "", regex=False),
        errors="coerce"
    )

# Preview the cleaned data
print("First 5 rows after cleaning:")
print(df.head())

print("\nDataFrame information:")
df.info()
```
